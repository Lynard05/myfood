```python
df_food = usda['Shrt_Desc']
df_g_serv = 100
df_Energ_Kcal = usda['Energ_Kcal']
df_Energ_Kcal_g = df_Energ_Kcal / df_g_serv
df_Energ_Kcal = pd.concat([df_food, df_Energ_Kcal, df_Energ_Kcal_g], axis = 1)
df_Energ_Kcal.columns = ['Food', 'Energ_Kcal', 'Energ_Kcal / g']
df_Energ_Kcal = df_Energ_Kcal.sort_values('Energ_Kcal / g', ascending = False).dropna()
df_Energ_Kcal[:20]
```

```python
df_protein = usda['Protein_(g)']
df_protein_g = df_protein / df_g_serv
df_protein = pd.concat([df_food, df_protein, df_protein_g], axis = 1)
df_protein.columns = ['Food', 'Protein_(g)', 'Protein_(g) / g']
df_protein = df_protein.sort_values('Protein_(g) / g', ascending = False).dropna()
df_protein[:20]
```
